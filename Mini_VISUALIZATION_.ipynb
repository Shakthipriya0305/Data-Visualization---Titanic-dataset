# MINI DATA VISUALIZATION DASHBOARD ON TITANIC DATASET



```python
# Importing the libraries and dataset
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as ny

df = pd.read_csv("Titanic-Dataset.csv")
```

# CLEAN THE DATASET

```python
# Fill missing Age values with the median
```

```python
df["Age"].fillna(df["Age"].median())
display(df["Age"])
```

```python
# Fill missing embarkment with the mode
```

```python
df["Embarked"].fillna(df["Embarked"].mode()[0],inplace=True)
display(df['Embarked'])
```

```python
# Drop the Cabin column
```

```python
df.drop(columns=["Cabin"],inplace=True)
```

# NEW FEATURE

```python
# Create new AgeGroup
```

```python
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0,12,18,30,50,80],
    labels=["child","teen","young adult","adult","senior"]
)
display(df["AgeGroup"])
```

```python
# Create family size
```

```python
df["FamilySize"]=df['SibSp']+df['Parch']
print(df['FamilySize'])
```

# VISUALIZATIONS

```python
# Histogram of Age Distribution
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=20,edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

The histograms shows the no. of people who boarded the Titanic are between 20 to 40.

```python
# Bar chart of Survival Rate
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Sex",y="Survived",data = df)
plt.title("Survival Rate by Sex")
```

Based on the graph,we can find that female have survived more than male by more than 5%.

```python
# Box plot on Passenger Fare
```

```python
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

Based on box plot,The median fare is highest for first-class passengers and lowest for third-class passengers. First class also shows the greatest variability in ticket prices due to several high-value outliers.

```python
# Scatter plot on Age Vs Fare
```

```python
sns.scatterplot(
    x="Age",
    y="Fare",
    hue="Survived",
    data=df,

)

plt.title("Age vs Fare (Colored by Survival)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

The scatter plot shows that there is no strong relationship between a passenger's age and the fare paid. However, passengers who paid higher fares appear to have a higher survival rate, while most lower-fare passengers are concentrated in the non-survived group.

```python
# Correlation Heatmap
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

The heatmap highlights the correlation between numerical features. Fare has a positive correlation with survival, while passenger class (Pclass) has a negative correlation with survival, indicating that first-class passengers were more likely to survive than those in lower classes
